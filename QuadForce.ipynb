```python
import pandas as pd
# For reading CSV file

df = pd.read_csv('ArianaGrande.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

import numpy as np

# If Nan values exists we will drop them from our dataframe and join all other lyrics .

X="".join(df["Lyric"].dropna())

# print(X)

Y=X.lower()

print(Y)

# print(df.shape)
```

```python
# For counting each word frequency.

d={}
for i in Y.split():
  if i in d:
    d[i]+=1
  else:
    d[i]=1
print(d)


```

```python
# Now , we sort the dictionary from most frequent word to least frequent word according to Zipf's Law statement.

d2 = sorted(d.items(), key=lambda item: item[1], reverse=True)


print(d2)

# for i , j in d2[:]:

  # print(f"{i}:{j}")


```

```python
word_rank=np.arange(1,len(d2)+1)
word_freq=[]

for i,j in d2:
  word_freq.append(j)

word_freq=np.array(word_freq)

print(word_freq)

print(word_rank)

# print(word_freq/ word_rank)
```

```python
# Plot Zipf's Law Curve(Actual and Expected)

plt.figure(figsize=(10,6))

plt.plot(word_rank,word_freq,label=" Actual Word Distribution ")


# We, know that frequency of any word is inversely proportional to its rank according to Zipf's Law Statement.

plt.plot(word_rank,word_freq[0]/ word_rank,label="Actual Zipf's Law Distribution")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Word-Rank")
plt.ylabel("Word-Frequency")

plt.legend()
plt.title("Zipf's Law - Word Frequency Distribution")

plt.show()
```

```python
# Data Visualization of Songs by Year

counts = df['Year'].value_counts().sort_index()
# print(counts)
plt.bar(counts.index.astype(str), counts.values)
plt.xlabel("Year")
plt.ylabel("Number of Songs")
plt.title("Songs Released Per Year")
plt.xticks(rotation=45)
plt.show()
```
